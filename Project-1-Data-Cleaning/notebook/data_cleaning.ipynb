# Load Data and Project Introduction
In this section, we read the dataset from the `data/` folder and prepare the DataFrame for analysis and cleaning.

```python
import pandas as pd
from pathlib import Path

# Use a relative path from the notebook folder to the data file.
data_path = Path("..") / "data" / "Dataset for Data Analytics.xlsx"

# Read the Excel file using the dependency listed in requirements.txt.
df = pd.read_excel(data_path, engine="openpyxl")

# Display the first 5 rows to verify successful loading.
df.head()
```

## Data Structure Check
Use `df.info()` to get an overview of columns, data types, and non-null counts.

```python
df.info()
```

## Missing Values Check
Count the missing values in each column to identify data gaps before cleaning.

```python
df.isnull().sum()
```

## Duplicate Check
Look for duplicate rows and duplicate `OrderID` values to validate dataset quality.

```python
# Number of fully duplicated rows
df.duplicated().sum()
```

```python
# Number of duplicate OrderIDs
df["OrderID"].duplicated().sum()
```

## Unique Customer Count
Check how many distinct customers appear in the dataset using `CustomerID`.

```python
df["CustomerID"].nunique()
```

## Business Logic Validation
Compare `Quantity * UnitPrice` with `TotalPrice`, rounding values to avoid floating-point precision issues.

```python
# Count rows where totals do not match after rounding to 2 decimals
(
    round(df["Quantity"] * df["UnitPrice"], 2)
    != round(df["TotalPrice"], 2)
).sum()
```

## Invalid Value Check
Verify that `Quantity` and `UnitPrice` are greater than zero, since non-positive values are not valid for orders.

```python
(df["Quantity"] <= 0).sum()
```

```python
(df["UnitPrice"] <= 0).sum()
```

## Review Mismatched Totals
If any rows do not match, display the first ten for manual inspection.

```python
invalid_total = df[df["Quantity"] * df["UnitPrice"] != df["TotalPrice"]]

invalid_total[
    ["Quantity", "UnitPrice", "TotalPrice"]
].head(10)
```

## Handle Missing Coupon Codes
Fill blank values in the `CouponCode` column with `No Coupon` because missing values likely mean no coupon was used.

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

## Verify After Cleaning
Recalculate missing values after cleaning the coupon column to confirm the cleanup is complete.

```python
df.isnull().sum()
```

## Save the Cleaned Data
Export the cleaned dataset to a new Excel file inside the `cleaned_data/` folder.

```python
output_path = "../cleaned_data/cleaned_dataset.xlsx"

df.to_excel(output_path, index=False)

print("Cleaned dataset saved successfully!")
```
